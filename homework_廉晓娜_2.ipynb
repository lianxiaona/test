## EP:
- 通过math库，写一个程序，使得用户输入三个顶点(x,y)返回三个角度
- 注意：Python计算角度为弧度制，需要将其转换为角度
<img src="../Photo/18.png">

```python
import math
x1,y1,x2,y2,x3,y3= eval(input('>>'))#先求出三角形的三条边
a = math .sqrt((x2-x3) ** 2 + (y2-y3) ** 2)
b = math .sqrt((x3-x1) ** 2 + (y3-y1) ** 2)
c = math .sqrt((x2-x1) ** 2 + (y2-y1) ** 2)
A = math .acos((a * a - b * b - c * c) / (-2 * b * c))#求三角形的每个角度
B = math .acos((b * b - a * a - c * c) / (-2 * a * c))
C = math . acos((c * c - b * b - a * a) / (-2 * a * b))
print(math.degrees(A),math.degrees(B),math.degrees(C))
```

## EP:
- 利用ord与chr进行简单邮箱加密

```python
mail = '2798366323@qq.com' #MD5摘要算法
for i in mail:
    print(chr(ord(i) + 10),end='')
```

```python
mail = '2798366323@qq.com' #MD5摘要算法
result = ''
for i in mail:
    result +=  chr(ord(i) + 10)
print(result)
```

```python
mail = '2798366323@qq.com' #MD5消息摘要算法
result = ''
for i in mail:
    result =  result + chr(ord(i) + 10)
    print(result)
print(result)
```

```python
import hashlib
#带加密信息
#创建MD5对象
str_ = 'This is a MD5 test'
h1 = hashlib.md5()
h1.update(str_.encode(encoding='utf-8'))
print('MD5加密前为：' + str_)
print('MD5加密后为：' + h1.hexdigest())
```

```python
import hashlib
dataset = 'eb67ac9650e6da8b57e2b79687038c73'
password = input('>>') 
h1 = hashlib.md5()
h1.update(password.encode(encoding='utf-8'))
if dataset == h1.hexdigest():
    print('登录成功')
else:
    print('登录失败')
    
```
