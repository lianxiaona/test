# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
#import randiant
# a = 10
# b = 15
# c = random.randint(a,b)
num = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','s','y','z']
random4 = random.randint(0,25)
print(num[random4])
random1 = random.randint(0,25)
print(num[random1])
random2 = random.randint(0,25)
print(num[random2])
random3 = random.randint(0,25)
print(num[random3])
while True:
    n,n1,n2,n3 = eval(input('请输入四个字母：'))
    if num[random4] == n and num[random1] == n1 and num[random2] == n2  and num[random3] == n3:
        print('你真厉害，猜对了')
        break
    else:
        print('请重新输入')
```

```python
import random
i=0
while i < 3:
    yanzhengma = str(random.randint(1000,9999))#将数字转换成字符串
    print('验证码是：',yanzhengma)
    input_ = input('>>')
    if yanzhengma == input_:
        print('验证码正确')
        break
    else:
        print('验证码错误，请再次输入')
        i += 1
else:
    print('验证码次数超限')
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):#默认从0开始，步长为1
    for j in range(10):
        for k in range(10):
             print(i,j,k)#这种多级嵌套非常耗时
```
